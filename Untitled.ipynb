```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

df=pd.read_csv('HRDataset.csv')

df.isnull()

df.drop(columns=['LastPerformanceReview_Date','DaysLateLast30'],axis=1,inplace=True)
df.dropna(thresh=2,inplace=True)
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')


#PODPUNKT1


perfscore_manname=sns.jointplot(x='PerformanceScore',y='ManagerName',data=df,height=15)

#Nie istnieje zależnośc pomiędzy tym, kto jest bezpośrednim przełożonym danego pracownika, a oceną wydajności pracy.

#PODPUNKT2

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2022,10,25)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')


df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)

seniority_recruitsource=sns.jointplot(x='Seniority',y='RecruitmentSource',data=df)

"""Nie istnieje złoty środek jeśli chodzi o sposób zatrudnienia pracownika, gwarantujący jego długi pobyt w firmie. 
Mimo to, w tej chwili najdluższym stażem pracy mogą pochwalić się osoby, które znalazły pracę dzięki ogłoszeniu na billboardzie."""

#PODPUNKT3

sns.boxplot(x=df['MaritalDesc'],y=df['EmpSatisfaction'],data=df)


#Największą satysfakcję z pracy mają wdowy(zadowolenie z pracy nie jest niższe niż 4 w 5-stopniowej skali. Istnieją przypadki, w których zadowolenie z pracy wynosi 5/5.)
#Najniższym stopniem zadowolenia z pracy charakteryzują się pracownicy będący singlami bądź żyjący w separacji ze swoim partnerem.

#PODPUNKT4

def age(row):
    birth = row['DOB']
    return (dt.datetime(2022,10,25)-row['DOB'])/np.timedelta64(1,'Y')
    

df['Age'] = df.apply(lambda row: age(row),axis=1)

df.drop(df[df['Age']<0].index, inplace=True)
df.drop(df[df['EmploymentStatus']!='Active'].index, inplace=True)

age_employmentstatus=sns.jointplot(x='Age',y='EmploymentStatus',data=df,height=10)

#Największe grono pracowników z aktywnym statusem zatrudnienia mieści się w przedziale wiekowym 33-41 lat.

#PODPUNKT5

age_specialprojects=sns.jointplot(x='Age',y='SpecialProjectsCount',data=df,height=10)

#Największą ilośc projektów specjalnych wykonują pracownicy w okolicach 35 r.ż.. Ilość wykonywanych przez pracowników projektów specjalnych spada wraz z starzeniem się pracownikow.
           
           

```
